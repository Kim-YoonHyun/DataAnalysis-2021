```python
import pandas as pd
df = pd.read_csv('해외여행.csv')
df
```

```python
text_content = ' '.join(df['내용'])
text_content
```

```python
text_title = ' '.join(df['제목'])
text_title
```

```python
text = text_content + ' ' + text_title
text
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
tokens = okt.nouns(text)
tokens[:10]
```

```python
# 영문자, 숫자 제거
import re
new_tokens = []
for token in tokens:
    item = re.sub('[A-Za-z0-9]', '', token)
    if item:
        new_tokens.append(item)
```

```python
# 한글 폰트 사용
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic') 
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
trip = nltk.Text(new_tokens, name='해외여행')
plt.figure(figsize=(15,6))
trip.plot(50)
plt.show()
```
