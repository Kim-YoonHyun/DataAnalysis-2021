### 반드시 런타임 다시 시작

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
# %matplotlib inline : 주피터 노트북에서 그래트를 볼 때 사용하는 매직 명령어
```

```python
plt.title('라인 플롯')
plt.plot([1, 4, 9, 16])
plt.grid()
plt.show()
```

```python
plt.title('line plot')
xs = np.linspace(0, 3, 301)
ys = (xs + 1) ** 2
plt.plot([1, 4, 9, 16])
plt.grid()
plt.show()
```

## 스타일 지정

```python
# 색깔(r), 마커(s)
plt.title("'rs--' style plot ")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.grid()
plt.show()
#
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("style example")
plt.show()
```

### 그림 범위 지정

```python
plt.title("x, y range setting")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-2, 20)
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("tick label setting")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

### 여러개의 선을 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("drawing more lines")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

### 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend plot")
plt.plot(X, C, ls="--", label="cosine")   # 라벨을 지정해야 범례에 표시됨
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.grid()
plt.show()
```

### x축, y출 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.grid()
plt.show()
```
